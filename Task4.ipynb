```python
# Mobile recharge validation system

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    elif amount not in valid_plans:
        print("Invalid plan selected.")
        print("Valid plans are:", valid_plans)
        return False
    else:
        print("Recharge successful!")
        return True


# Main program with retry option
while True:
    try:
        recharge_amount = int(input("Enter recharge amount: ₹"))
        
        if validate_recharge(recharge_amount):
            break   # Exit loop if recharge is valid
        
        print("Please try again.\n")
        
    except ValueError:
        print("Please enter a valid numeric amount.\n")
```

```python
# Inventory reorder alert system

def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert (Stock = {quantity})")
        else:
            print(f"{product}: Stock OK (Stock = {quantity})")


# Sample inventory dictionary
inventory = {
    "Rice": 25,
    "Sugar": 10,
    "Oil": 18,
    "Wheat": 12,
    "Salt": 30
}

# Function call
check_inventory(inventory)
```

```python
# Student result processing system

def process_result(marks):
    total = 0
    
    # Calculate total using loop
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    # Determine result
    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"


# Example student marks list
student_marks = [60, 45, 55, 70, 40]

# Function call
avg, result = process_result(student_marks)

print("Average Marks:", avg)
print("Result:", result)
```

```python
# Cab fare estimator with retry option

def calculate_fare(distance, peak):
    base_fare = 50
    rate_per_km = 12
    
    fare = base_fare + (distance * rate_per_km)
    
    if peak:
        fare += fare * 0.25   # 25% extra during peak hours
    
    return fare


# Main program with retry option
while True:
    try:
        distance = float(input("Enter distance in km: "))
        
        if distance <= 0:
            print("Distance must be greater than 0.\n")
            continue
        
        peak_input = input("Is it peak hour? (yes/no): ").lower()
        peak_hour = True if peak_input == "yes" else False
        
        total_fare = calculate_fare(distance, peak_hour)
        
        print(f"Total Cab Fare: ₹{total_fare:.2f}")
        
        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            print("Thank you for using the Cab Fare Estimator!")
            break
            
    except ValueError:
        print("Please enter valid numeric input.\n")
```

```python
# Employee attendance eligibilty checker

def check_eligibility(attendance_list):
    present_days = 0
    
    # Count Present days using loop
    for status in attendance_list:
        if status == "P":
            present_days += 1
    
    total_days = len(attendance_list)
    attendance_percentage = (present_days / total_days) * 100
    
    # Determine eligibility
    if attendance_percentage >= 75:
        return attendance_percentage, "Eligible"
    else:
        return attendance_percentage, "Not Eligible"


# Example attendance record
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

# Function call
percentage, result = check_eligibility(attendance)

print(f"Attendance Percentage: {percentage:.2f}%")
print("Status:", result)
```
